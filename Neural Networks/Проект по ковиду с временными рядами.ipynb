### Импортируем все необходимое

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn import preprocessing
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings('ignore')
```

### Загружаем данные и обрабатываем их

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data.head()
```

```python
#test_data.head()
```

```python
le = LabelEncoder()
le.fit(test_data.Country_Region)
test_data.Country_Region = le.transform(test_data.Country_Region)

le = LabelEncoder()
le.fit(test_data.Province_State)
test_data.Province_State = le.transform(test_data.Province_State)
```

```python
test_data['Date'] = pd.to_datetime(test_data['Date'], infer_datetime_format=True)
test_data.rename({'ForecastId':'Id'}, axis=1, inplace=True)
test_data
```

```python
le = LabelEncoder()
le.fit(train_data.Country_Region)
train_data.Country_Region = le.transform(train_data.Country_Region)

le = LabelEncoder()
le.fit(train_data.Province_State)
train_data.Province_State = le.transform(train_data.Province_State)
```

```python
train_data['Date'] = pd.to_datetime(train_data['Date'], infer_datetime_format=True)
train_data
```

```python
X = train_data.drop(['ConfirmedCases', 'Fatalities'], axis=1)
y = train_data[['ConfirmedCases', 'Fatalities']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3)
```

```python
y_train_cases = y_train.ConfirmedCases
y_train_fatal = y_train.Fatalities

y_val_cases = y_val.ConfirmedCases
y_val_fatal = y_val.Fatalities
```

### CatBoost (ConfirmedCases)

```python
from catboost import CatBoostRegressor

cat_model = CatBoostRegressor()
cat_model.fit(X_train, y_train_cases)
```

```python
cat_model.score(X_val, y_val_cases)
```

```python
y_test_cases = cat_model.predict(test_data)
```

```python
y_test_cases
```

### CatBoost (Fatalities)

```python
cat_model = CatBoostRegressor()
cat_model.fit(X_train, y_train_fatal)
```

```python
cat_model.score(X_val, y_val_fatal)
```

```python
y_test_fatal = cat_model.predict(test_data)
```
